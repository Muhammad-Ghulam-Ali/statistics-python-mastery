```python
# ── COMPLEMENT RULE ────────────────────────────────────────
p_default = 0.05
p_no_default = 1 - p_default

print("COMPLEMENT RULE")
print(f"P(default):     {p_default}")
print(f"P(no default):  {p_no_default}")
```

```python
# ── ADDITION RULE ──────────────────────────────────────────
# Mutually exclusive — borrower is either high risk or low risk
p_high_risk = 0.40
p_low_risk = 0.40
p_either = p_high_risk + p_low_risk

print("\nADDITION RULE — Mutually Exclusive")
print(f"P(high or low risk): {p_either}")
```

```python
# Not mutually exclusive — female OR from Lahore
p_female = 0.50
p_lahore = 0.30
p_both_fl = 0.15
p_female_or_lahore = p_female + p_lahore - p_both_fl

print("\nADDITION RULE — Not Mutually Exclusive")
print(f"P(female or Lahore): {p_female_or_lahore}")
```

```python
# ── MULTIPLICATION RULE ────────────────────────────────────
# Two independent borrowers, each 20% chance of default
p_default_one = 0.20
p_both_default = p_default_one * p_default_one

print("\nMULTIPLICATION RULE")
print(f"P(both default): {p_both_default}")
```

```python
# ── CONDITIONAL PROBABILITY ────────────────────────────────
p_missed = 0.20
p_default_and_missed = 0.04
p_default_given_missed = p_default_and_missed / p_missed

print("\nCONDITIONAL PROBABILITY")
print(f"P(default | missed payment): {p_default_given_missed}")
print(f"Default rate jumped from 5% to {p_default_given_missed*100:.0f}%")
print("Knowing they missed a payment changed everything.")
print("=" * 160)
```
