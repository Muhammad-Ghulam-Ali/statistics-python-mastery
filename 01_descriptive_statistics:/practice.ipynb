```python
import pandas as pd
import numpy as np
np.random.seed(42)
```

```python
repayment = np.concatenate([
    np.random.normal(loc=8000, scale=2000, size=190),
    np.random.normal(loc=80000, scale=10000, size=10)
])
```

```python
df = pd.DataFrame({
    "repayment": 
    repayment
})
df
```

```python
# MEASURE OF CENTRAL TENDENCIES

mean_val = df['repayment'].mean()
median_val = df['repayment'].median()
mode_val = df['repayment'].mode()[0]
```

```python
# MEASURE OF DESPERSION
std = df['repayment'].std()
var = df['repayment'].var()
```

```python
# PRINTNG THE MEASURES
print("="*40)
print("MEASURES OF CENTERAL TENDENCY")
print("-" * 40)
print(f"Mean:                         PKR {mean_val:,.0f}")
print(f"Median:                       PKR {median_val:,.0f}")
print(f"Mode:                         PKR {mode_val:,.0f}")
print(" "*40)
print("="*40)
print("MEASURES OF DEVIATION")
print("-" * 45)
print(f"Standard Deviation:            PKR {std:,.0f}")
print(f"Variance:                      PKR {var:,.0f}")
```

**PROBLEM/PRACTICE**

```python
income = np.array ([25000, 30000, 28000, 35000, 32000, 29000,
27000, 31000, 26000, 500000, 33000, 30000])
df = pd.DataFrame({
    "income": income
})
df
```

```python
# Centeral tendencies
mean = df['income'].mean()
median = df['income'].median()
mode = df['income'].mode()[0]

# Despersion
variance = df['income'].var()
standard_dev = df['income'].std()
```

```python
print("="*40)
print("MEASURES OF CENTERAL TENDENCY")
print("-" * 40)
print(f"Mean:                         PKR {mean:,.0f}")
print(f"Median:                       PKR {median:,.0f}")
print(f"Mode:                         PKR {mode:,.0f}")
print(" "*40)
print("="*40)
print("MEASURES OF DEVIATION")
print("-" * 50)
print(f"Standard Deviation:            PKR {standard_dev:,.0f}")
print(f"Variance:                      PKR {variance:,.0f}")

# See the impact of outliers clearly
print(" " * 50)
print("=" * 50)
print("GAP")
print(f"PKR {(mean - median):,.0f}")
```
